```python
# Imports

import os
import numpy as np
import pandas as pd
import deepchem as dc
```

```python
# Read test and train data

root_dir = '/Users/HQ/Repositories/DILI/'
data_dir = os.path.join(root_dir, 'Data')

test_data  = pd.read_csv(os.path.join(data_dir, "Test_data_features.csv"))
train_data = pd.read_csv(os.path.join(data_dir, "Train_data_features.csv"))

train_data = train_data[~train_data.InChIKey14.isin(test_data.InChIKey14.to_list())].reset_index(drop=True)

y_dili = np.array(test_data['TOXICITY'])
test_features= test_data.drop(["TOXICITY", "smiles_r", "Source_rank", "Source", "Data"] , axis = 1)
```

```python
# Split

smiles = test_data.smiles_r  # your list of SMILES strings
activities = test_data.TOXICITY  # your list of activities or properties associated with each SMILE
dataset = dc.data.DiskDataset.from_numpy(X=activities, ids=smiles)

# Initialize the ButinaSplitter
splitter =dc.splits.ButinaSplitter(cutoff = 0.7)

# Split the dataset
train_dataset, test_dataset = splitter.train_test_split(dataset, frac_train=0.80, seed=42)

ncv_smiles  = train_dataset.ids
heldouttest_smiles = test_dataset.ids

test_data_ncv = test_data[test_data.smiles_r.isin(ncv_smiles)]
test_data_ncv = test_data_ncv.reset_index(drop=True)
test_data_heldouttest = test_data[test_data.smiles_r.isin(heldouttest_smiles)]
test_data_heldouttest = test_data_heldouttest.reset_index(drop=True)

test_data_ncv.to_csv(os.path.join(data_dir, 'test_data_ncv_DILIst_888.csv'))
test_data_heldouttest.to_csv(os.path.join(data_dir, 'test_data_heldouttest_DILIst_223.csv'))

smiles_list = test_data_ncv.smiles_r.to_list()
with open(os.path.join(data_dir, 'test_data_ncv_smiles_list.txt'), 'w') as f:
    for smiles in smiles_list:
        f.write("%s\n" % smiles)

smiles_list = test_data_heldouttest.smiles_r.to_list()
with open(os.path.join(data_dir, 'test_data_heldouttest_smiles_list.txt'), 'w') as f:
    for smiles in smiles_list:
        f.write("%s\n" % smiles)

test_data.to_csv(f'{data_dir}/Test_data_DILIst_1111.csv', index=False)
train_data.to_csv(f'{data_dir}/Train_data_liv_tox_14256.csv', index=False)
```
